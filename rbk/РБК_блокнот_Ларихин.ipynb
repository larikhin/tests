					
Наглядно представить ТОП-20 материалов по просмотрам за неделю						
Оформить ТОП-20 в виде таблицы с заголовками материалов, при этом заголовок необходимо сделать активной ссылкой с возможностью перехода на статью на сайте (адрес страницы - столбец URL на листе База публикаций).						
Визуализировать полученные данные						
Все этапы выполнения задания должны быть сохранены в файле или описаны отдельно						
						
Обратите внимание: у каждой публикации есть уникальный id, который находится в конце адреса страницы (например, 5c4af9ed9a7947bd94fe0cc9), при этом полный адрес страницы может отличаться						
						
Из отчёта должны быть исключены просмотры, содержащие след. URL: 						
test.pro						
test.v2.pro						
feature-rbcnews						
/preview/						
staging.pro						
staging.v2.pro						
						
Лист База публикаций - список опубликованных материалов со ссылками на сайт и заголовками						
Лист Выгрузка - данные по просмотрам материалов за отчётный период						

```python
import pandas as pd

publication = pd.read_csv('База_публикаций.csv', delimiter=',', decimal='.')
loadings = pd.read_csv('Выгрузка.csv', delimiter=',', decimal='.')
```

```python
# удаление из отчёта просмотры, содержащие след. URL:
# test.pro 
# test.v2.pro
# feature-rbcnews
# /preview/
# staging.pro
# staging.v2.pro 

loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("test.pro")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("test.v2.pro")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("feature-rbcnews")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("/preview/")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("staging.pro")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("staging.v2.pro")== False ]
loadings = loadings[loadings["Отчет за период с 2019-02-11 по 2019-02-17"].str.contains("Адрес страницы")== False ]
```

```python
# Выборка топ 20 публикаций по количеству просмотров
loadings_top_20 = loadings.iloc[0:20]
loadings_top_20.reset_index(drop= True , inplace= True )
loadings_top_20.index+=1
loadings_top_20
```

```python
# формирование таблицы "ID публикаций"
loadings_strip = loadings_top_20.copy(deep=True)

def clean_data(response):
    x = response.split('/')[-1]
    return x

loadings_strip['Отчет за период с 2019-02-11 по 2019-02-17'] = loadings_strip["Отчет за период с 2019-02-11 по 2019-02-17"].apply(lambda x: clean_data(x)) 
loadings_strip.rename(columns = {'Отчет за период с 2019-02-11 по 2019-02-17':'ID публикации','Unnamed: 1':'Просмотры'}, inplace = True )
loadings_strip
```

```python
import numpy as np
def change_type(x):
    return int(x)

# объединение двух таблиц в одну
merge_total = publication.merge(loadings_strip, how='inner')
#merge_total =merge_total.astype({'Просмотры': np.int64})
merge_total['Просмотры'] = merge_total['Просмотры'].apply(lambda x: change_type(x))
merge_total = merge_total.sort_values(by = 'Просмотры', ascending=False)
merge_total.reset_index(drop= True , inplace= True )
merge_total.index+=1

merge_total
```

```python
# создания итогового файла Google Таблицы с гиперссылками ТОП20 статей
df_total = pd.DataFrame()
# =hyperlink("www.google.com";"Поиск в Интернете")
df_total["ТОП 20"] = '=hyperlink("' + merge_total['URL'] + '";"' + merge_total['Заголовок материала'] + '")'
df_total["Просмотры"] = merge_total['Просмотры']
df_total.to_excel(r'top.xlsx')
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(10, 6))
plt.bar(merge_total['Заголовок материала'], merge_total['Просмотры'],color='grey', width=0.3, align='center')
plt.title('Лидеры просмотров')
plt.xlabel("Номер")
plt.ylabel("Количество просмотров")
plt.xticks(rotation = 'vertical')
plt.grid(True)
plt.show()
```
