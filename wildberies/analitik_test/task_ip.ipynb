```python
from pandas import Series, DataFrame
import pandas as pd

pd.set_option('max_columns', None)
sku=pd.read_csv('2.csv', delimiter=',', decimal=',')
```

```python
'''
1. Выгрузка в Exel c Mpstats 
ссылка на выгрузку https://disk.yandex.ru/d/QT8LT41yNo791A : 
профильтровать все SKU по параметрам: 
+средний чек не ниже 1500р., 
+выручка более 1 млн рублей, 
+комментариев не менее 50. 
+Убрать ООО из выдачи. 
+Оставить только ИП.

2. Найти ТОП 30 поставщиков 
по «купальникам» и 
составить таблицу с критериями выбора, 
чтобы отдать в отдел закупок

3. Описать методологию поиска поставщиков 
(как понял задание пусть так и делает).
 Где искать поставщиков - его вопрос. 
Посмотрим на логику действий.'''
```

```python
# средний чек не ниже 1500р., 
s1 = sku[sku['Average price']>=1500]
s2 = s1[s1['Revenue']>=1000000]
s3 = s2[s2['Comments']>=50]
s4 = s3.loc[s3.Seller.str.contains ("ИП")]
s5 = s4.drop(['URL'], axis=1)
s5 = s5.drop(['Thumb'], axis=1)
s6 = s5[s5['Lost profit']==0]
s6 = s6[s6['Rating']==5]
s6 = s6.sort_values(['Sales Per Day Average'], ascending=[False])
s6
```

```python
s6.to_csv('filtered_file.csv')
```

```python
s7 = DataFrame(s6, columns=['Name','Category','Brand','Color'])
s7 = s7.groupby(['Brand','Color'], as_index= True ).agg({'Name': ' '. join })

s7.to_csv('list_for_purchase.csv')
```
